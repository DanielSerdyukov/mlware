```python
import matplotlib.pyplot as plot
plot.style.use('seaborn-whitegrid')
import numpy as np
from scipy import interpolate
```

```python
def mov_avg(seq, window=3):
    s = np.insert(np.cumsum(seq), 0, [0])
    return (s[window :] - s[:-window]) * (1. / window)
```

```python
x = np.linspace(0, 100, 100)
y = np.random.normal(size=len(x))
yMA = mov_avg(y)
xMA = x[len(x)-len(yMA):]
```

```python
raw = plot.figure(1, figsize=(16, 10))
plot.plot(x, y, 'r-')
plot.plot(xMA, yMA, 'b-', lw=2)
spl = interpolate.BSpline(xMA, yMA, k=3)
xSp = np.linspace(xMA.min(), xMA.max(), 1000)
plot.plot(xSp, spl(xSp), 'k-', lw=3)
plot.ylim(y.min(), y.max())
# plot.savefig("dataset/raw.png")
plot.show()
```

```python
xmax = np.argmax(y)
lx = max(0, xmax - 10)
rx = min(len(x), xmax + 10)
print(lx, xmax, rx)
print(x[lx:rx])
```
